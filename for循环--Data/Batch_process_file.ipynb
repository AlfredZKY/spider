```python
import pandas as pd 
import os
```

```python
# 读取指定的数据目录
path = r'./for循环--Data/datas'

# 构造一个空列表用于存储所有的文件数据
datas = []
for filename in os.listdir(path):
    # 判断文件格式
    if filename.find('txt') != -1:
        # 读取txt格式的数据
        data = pd.read_csv(path + '\\' + filename,sep = '\t')
    else:
        # 读取csv格式的数据
        data = pd.read_csv(path + '\\' + filename)
    # 将读取到的每一个数据集存储到datas列表中
    datas.append(data)
# 将所有的数据集进行合并
final_data = pd.concat(datas)
final_data.head()
```
