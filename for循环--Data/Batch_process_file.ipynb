```python
import pandas as pd 
import os
import jieba
import collections # 用于词频的排序
```

```python

path = r'./for循环--Data/datas'

# 构造一个空列表用于存储所有的文件数据
datas = []
for filename in os.listdir(path):
    # 判断文件格式
    if filename.find('txt') != -1:
        # 读取txt格式的数据
        data = pd.read_csv(path + '\\' + filename,sep = '\t')
    else:
        # 读取csv格式的数据
        data = pd.read_csv(path + '\\' + filename)
    # 将读取到的每一个数据集存储到datas列表中
    datas.append(data)
# 将所有的数据集进行合并
final_data = pd.concat(datas)
final_data.head()
```

数据单位的统一处理

```python
apps = pd.read_excel(r'./for循环--Data/apps.xlsx')

# 处理安装量字段install，将数据统一为以"万"为单位
install_new = []

for i in apps['install']:
    # 判断以"亿"为单位的数据
    if i.find("亿") != -1:
        install_new.append(str(float(i[:-1])*10000) + "万")
    elif i.find("万") != -1:
        install_new.append(i)
    else:
        install_new.append(str(int(i)/10000) + "万")

apps['install'] = pd.Series(install_new)

# 处理安装量字段size，将数据统一为以"MB"为单位
size_new = []

for i in apps['size']:
    # 判断以"亿"为单位的数据
    if i.find("KB") != -1:
        size_new.append(str(round(float(i[:-2])/1024,2)) + "MB")
    else :
        size_new.append(i)

apps['size'] = pd.Series(size_new)

apps.head()
```

词频统计

```python
# 读取新闻报道
fn = open('./for循环--Data/news.txt',encoding='utf-8')
news = fn.readlines()
# 将新闻报道的隔断内容拼接为一个整体
news = ''.join(news)

# 删除特殊字符，"\n"
news = news.replace('\n','')

# 对新闻内容切词，并保留一个长度以上的词
words = []
for i in jieba.lcut(news):
    # 通过逻辑判断，保留至少两个字符的词
    if len(i) > 1:
        words.append(i)
#将词频结果存储到字段中
wf = {}
for i in set(words):
    wf[i] = words.count(i)

# 对词频结果进行排序
collections.Counter(wf).most_common()
```
